```python
# task2
EncMsg='###!!@mocleW EPGTQ!!!6789'
def dec(Encmsg):
    corePart=''.join(c for c in Encmsg if c.isalpha() or c.isspace())
    splitted=corePart.split()
    splitted[0]=splitted[0][::-1]
    print(' '.join(splitted))

dec(EncMsg)
```

```python
# task 3
EncMsg2='&&&**$gnirtS PLIO!!@1234'
EncMsg3='##$$$@!yalpstcejorp EPUVT****9887'
def dec(Encmsg):
    corePart=''.join(c for c in Encmsg if c.isalpha() or c.isspace())
    splitted=corePart.split()
    splitted[0]=splitted[0][::-1]
    splitted[1]=splitted[1].replace('O','U').replace('I','E')
    print(' '.join(splitted))


dec(EncMsg2)


```
